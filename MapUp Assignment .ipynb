```python
pd.DataFrame(data)
```

```python
def get_type_count(data):
    data['car_type'] = data['car'].apply(lambda x:'Low' if x <=15 else 'Medium' if x<=25 else 'High')

#Finding type count 
    count_types=data['car_type'].value_counts().to_dict()

#Sorting the new dict

    count_types_sorted = dict(sorted(count_types.items())) 
    
     
    return count_types_sorted
```

```python
count_types= get_type_count(data)


print(count_types)
```

```python
def get_bus_indexes(data):
    
#To find mean of bus data
    mean = data['bus'].mean()
    
#Findind the indices

    bus_indexes = data[data['bus'].apply(lambda x: x > 2 * mean)].index.tolist()
    
#Sorting the data in asc 
    
    sorted_bus_indexes=bus_indexes.sort()
    return sorted_bus_indexes
```

```python
bus_indexes = get_bus_indexes(data)

print(bus_indexes)
```
